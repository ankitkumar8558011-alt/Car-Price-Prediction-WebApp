## Cleaning

```python
backup=df.copy()
```

```python
# only numeric values will stay
df=df[df['year'].str.isnumeric()] 
```

```python
# change the year type
df['year']=df['year'].astype(int)
df.info()
```

```python
# it will remove the row in which "ask for price" is writen
df=df[df['Price']!="Ask For Price"]
```

```python
df['Price']=df['Price'].str.replace(',','').astype(int)
```

```python
# we split this values 45,000 kms through comas and then get 45000 and repace the comas
df['kms_driven']=df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
df=df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven']=df['kms_driven'].astype(int)
```

```python
df.info()
```

```python
# remove the row in which NAN value is there
# ~(tilde) use to excludeing this row i need all other rows
df=df[~df['fuel_type'].isna()]
```

```python
df['name']=df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df.reset_index(drop=True)
```

```python
df.info()
df.describe()
```

```python
df=df[df['Price']<6e6].reset_index(drop=True)
```

```python
df.head()
```

```python
# we store the cleaned data into now csv file 
df.to_csv('cleaned car.csv')
```

# Model

```python
x=df.drop(columns='Price')
y=df['Price']
```

#### importing train_test_split
- train_test_split() is a Scikit-learn function used to divide your dataset into training and testing data.

- It helps your Machine Learning model:

- Learn from training data
- Get evaluated on unseen testing data

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression # importing linear regression model
from sklearn.metrics import r2_score       # it tells how well the model is performing 
from sklearn.preprocessing import OneHotEncoder # it is used to convert categorical data into numerical data
from sklearn.pipeline import make_pipeline # it is used to make a pipeline
from sklearn.compose import make_column_transformer # it is used to apply different transformations to different columns
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_ 
# it will show the categories of each categorical column
# create the column transformer used to apply different transformations to different columns
```

```python
# create the column transformer used to apply different transformations to different columns
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough') 
```

```python
lr=LinearRegression() # create the linear regression model
```

```python
pipe=make_pipeline(column_trans,lr) # create the pipeline
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
y_pred
```

### r2_score
- R² score tells us how well our model explains the output data.
- It measures how accurate your regression model is
- It shows how much variation in the target variable is explained by the model

```python
r2_score(y_test,y_pred) # it tells how well the model is performing
```

```python
scores=[]
for i in range (1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred)) 
```

```python
np.argmax(scores) # the argmax gives the index of maximum value in the scores list
```

```python
scores[np.argmax(scores)]
```
